We extrat all the reference, we drop the number at this stage. The first reference in the follwing serie is we believe of importance. In futher analysis it appears that the main attribute matches with the current definition of the project. 

```python
back = pd.Series(str)
for index , value in dulux_remarks.items():
    back[index] = re.findall('\\*[A-Za-z]+\s*[A-Za-z]+', dulux_remarks[index])
```
